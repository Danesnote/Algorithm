<strong><font color="blue" size="3em">연습1: 리스트 변수로 큐를 다루는 enqueue, dequeue 기능 구현해보기</font></strong>
</div>

```python
queue_list = []

def enqueue(data):
  queue_list.append(data)

def dequeue():
  data = queue_list[0]
  del queue_list[0]
  return data
```

```python
for index in range(10):
  enqueue(index)
```

```python
len(queue_list)
```

```python
dequeue()
```

연습2(책엔 없음): 리스트 변수로 Lifoqueue 만들어보기

```python
queue_list = []

def enqueue(data):
  queue_list.append(data)

def dequeue():
  n = len(queue_list)
  data = queue_list[n-1]
  del queue_list[n-1]
  return data
```

```python
for index in range(10):
  enqueue(index)
```

```python
len(queue_list)
```

```python
dequeue()
```
